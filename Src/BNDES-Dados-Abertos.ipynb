# Imports

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

# Regi√£o Norte

```python
# Dados

norte = pd.read_csv("C:\\Users\\RenanSardinha\\Documents\\Data Science\\Projects\\BNDES-Dados-Abertos\\Dados\\regiao-norte.csv")
norte
```

```python
# Correla√ß√£o

print(norte.corr())
sns.heatmap(norte.corr())
plt.title('Regi√£o Norte', fontsize=14)
```

```python
# Exluir os valores faltantes

regiao_norte = norte.copy()
regiao_norte = regiao_norte.dropna()
```

```python
# M√©todos dos M√≠nimos Quadrados (MMQ)

X = regiao_norte.iloc[:, :-1]
y = regiao_norte.iloc[:,-1]

X = sm.add_constant(X)
model = sm.OLS(y,X) 
results = model.fit()
print(results.summary())
```

```python
# Regress√£o Linear 

X = regiao_norte['Valor Desembolsado (Reais)'].values.reshape(-1,1)
y = regiao_norte['IDH'].values.reshape(-1,1)

reg = LinearRegression()
reg.fit(X, y)


print("O modelo de IDH da Regi√£o Norte √© = {:.5} + {:.5}X".format(reg.intercept_[0], reg.coef_[0][0]))
```
