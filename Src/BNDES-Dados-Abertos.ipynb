## Bibliotecas

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

### Valores Desembolsados para Financiamentos de Operações não automáticas por Regiões

```python
regioes = pd.read_csv("C:\\Users\\RenanSardinha\\Documents\\Data Science\\Projects\\BNDES-Dados-Abertos\\Dados\\valor-desembolsado-por-regiao.csv")
regioes
```

```python
regiao = ['Sudeste', 'Nordeste', 'Sul', 'Norte', 'Centro-Oeste']
porcentagem = [42.68, 21.78, 12.38, 12.27, 10.89]
plt.figure(figsize = (16,7))
plt.bar(regiao, porcentagem)

plt.title("Valores Desembolsados para Financiamentos de Operações não automáticas, no período de 2013 - 2017", fontsize=16)
plt.xlabel("Regiões", fontsize=14)
plt.ylabel("Porcentagem", fontsize=14)
plt.show()
```

## Região Norte

```python
# Dados

norte = pd.read_csv("C:\\Users\\RenanSardinha\\Documents\\Data Science\\Projects\\BNDES-Dados-Abertos\\Dados\\regiao-norte.csv")
norte
```

```python
# Correlação

print(norte.corr())
sns.heatmap(norte.corr())
plt.title('Região Norte', fontsize=14)
```

```python
# Exluir os valores faltantes

regiao_norte = norte.copy()
regiao_norte = regiao_norte.dropna()
```

```python
# Métodos dos Mínimos Quadrados (MMQ)

X = regiao_norte.iloc[:, :-1]
y = regiao_norte.iloc[:,-1]

X = sm.add_constant(X)
model = sm.OLS(y,X) 
results = model.fit()
print(results.summary())
```

```python
# Regressão Linear 

X = regiao_norte['Valor Desembolsado (Reais)'].values.reshape(-1,1)
y = regiao_norte['IDH'].values.reshape(-1,1)

reg = LinearRegression()
reg.fit(X, y)


print("O modelo de IDH da Região Norte é = {:.5} + {:.5}X".format(reg.intercept_[0], reg.coef_[0][0]))
```
